#### - student-py



````python
## student.py

import json

"""
JSON formatet i graderegs.txt til reg af student, fag & karakter i form

{
    "students": {
        "jesper hansen": {
            "grades": {
                "matematik": 7,
                "engelsk": 12,
                "dansk": 7
            }
        }
    }
}


"""


# GradeReg Klasse

class GradeReg:
    """
    "Klasse GradeReg til håndtering af studerendes karakterer i forskellige fag på 7-trinsskalaen"

    Methods:
    __init__(*gradefile) - init af klassen med navn på din karakter fil(skal være oprettet) *gradefile er optionel. Default ="graderegs.txt"
    check_grade(grade) - tjekker om en given karakter er gyldig på skalaen
    new_entry(student, grade, class_name) - tilføjer en ny karakter til en studerende i et givent fag
    max_grades() - returnerer navnet på den studerende med den højeste totale karakter og gennemsnittet af karaktererne.
    load_grades() - indlæser karakterdata fra filen i JSON format
    save_grades() - gemmer karakterdata i filen i JSON format
    str() - returnerer karakterdata i JSON format (til debugging)
    
    """


    def __init__(self, gradefile="graderegs.txt"):
        self.gradefile = gradefile
        self.grade_range = [-3, 0, 2, 4, 7, 10, 12] # 7 Trinsskala - 0(00) & 2(02), tages der højde for der castes til int
        self.entries = self.load_grades() #loader entries fra vores gradefile
        
    def check_grade(self, grade:int): # obs: hvis check_grade skal bruges seperat, bør try/except integreres og logikken skrives om fra new_entry
        
        # tjekker om karakteren er i self.grade_range 
        return grade in self.grade_range
        
        
    def new_entry(self, student, grade, class_name):

        student = student.lower() # sikre at alt er lowercase, så der ikke kommer duplikates
        class_name = class_name.lower()
        try:
            #vi forsøger at caste grade til en integer i vores try/except
            grade = int(grade)
            
        except ValueError:
            
            return "Fejl: Karakteren skal være numerisk"
        # tjekker om karakteren er gyldig
        if not self.check_grade(grade):
            return "Fejl: Karakter er ikke på 7 trinsskalaen"


        # tjekker om studenten er oprettet eller faget er oprettet(ellers oprettes det)
        if student not in self.entries["students"] or class_name not in self.entries["students"][student]["grades"]:
            self.entries["students"][student] = {"grades": {class_name: grade}}
            self.save_grades()
            return f"{student.capitalize()} er oprettet med karakteren {grade} i {class_name.capitalize()}"
        else:
            return f"{student.capitalize()} har allerede fået en karakter i {class_name.capitalize()}"


    # laver en list med studenten og hans totale karakterer og gennemsnit
    def max_grades(self):
        student_list = []
        #loop igennem entries dict
        for student in self.entries["students"]:
            total_grade = sum(self.entries["students"][student]["grades"].values()) #Sum af alle grades values under den pågældende student
            student_list.append([student, total_grade, total_grade / len(self.entries["students"][student]["grades"])]) # listen indeholder navn, total karakter og gennemsnit
            max_grade = max(student_list, key=lambda x: x[1]) #lambda til finde max i index 1 på listen
        
        # returnere max_grade i letlæselig format
        return f"{max_grade[0].capitalize()} har den højeste totale karakter med: {max_grade[1]}\nGennemsnit: {round(max_grade[2],1)}"

    def load_grades(self):
        # åbn gradefile og load i JSON formatet og returnere til vores entries dict
        try:
            with open(self.gradefile, "r") as file:
                return json.loads(file.read())
        except FileNotFoundError:
            return {"students": {}}
        
    # save_grades - bruger json.dumps til at dumpe vores dict til json format og indent gør den letlæselig hvis den åbnes 
    def save_grades(self):
        with open(self.gradefile, "w") as file:
            file.write(json.dumps(self.entries, indent=4))
    
    # printer letlæseligt vores entries i json format
    def __str__(self):
        return json.dumps(self.entries, indent=4)


````
