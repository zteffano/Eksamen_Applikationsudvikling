# Case 1 - Trekant Applikation

Hvis du får tre længder, kan du eller måske ikke være i stand til at arrangere dem i en trekant. 
Hvis en af længder for eksempel er 10 cm lang, og de to andre er 3 cm lange, er det klart, at du 
ikke vil kunne få de korte længder til at mødes på midten. 
For alle tre længder er der en simpel test for at se, om det er muligt at danne en trekant: Hvis nogen af de tre længder er større end summen af de to andre, kan du ikke danne en trekant. Ellers kan du. 
Hvis summen af to længder er lige med den tredje, danner de det, der kaldes en "degenereret" trekant.
---

- Skriv en klasse ved navn Triangle, der tager tre heltal som attributter. 
- Skriv en funktion ved navn is_triangle, der udskriver enten "Ja" eller "Nej", alt efter om du kan eller ikke kan danne en trekant med de givne længder.
- Skriv en funktion, der beder brugeren om at indtaste tre længder, konvertere dem til heltal, og brug is_triangle til at kontrollere, om de givne længder kan danne en trekant.
- Skriv en funktion, der afregner trekantarealet.
- Skriv en funktion, der tegner trekanten og viser trekantens arialet på GraphWin ved brug af graphics.py modul.

### 4. Implementer designet



#### - Triangle.py



````python
## Triangle.py

from graphics import *
import math


# klasse Triangle

class Triangle:

    """
    Klassen Triangle repræsenterer en trekant med siderne a, b og c.

    Konstruktør:
    Initialiserer en trekant med siderne a, b og c. Input skal være numeriske værdier.

    Metoder:
    is_triangle(): Returnerer "JA" hvis trekanten er gyldig, ifølge formlen for en gyldig trekant, ellers "NEJ"
    area(): Returnerer trekantens areal, beregnet ved hjælp af Heron's formel.
    __str__(): Returnerer trekantens side længder som en string.
    find_third_point(x1, y1, x2, y2, a, b, c): Returnerer det tredje punkt i trekanten, givet de to andre punkter og de tre side længder.
    draw(): Tegner trekanten i et grafisk vindue ved hjælp af graphics.py objekter.

"""


    # Constructor - med 3 parametre side1, side2 og side3
    def __init__(self, a, b, c):
        try:
            self.a = int(a)
            self.b = int(b)
            self.c = int(c)
            self.normalizer = max(self.a, self.b, self.c) # normalizer bruges til at tegne trekanten i vinduet, uden at den bliver for lille eller stor
        except:
            raise ValueError("Input skal være et tal")

    # is_triangle method skal returnere True hvis trekanten er gyldig, ift. formlen for en gyldig trekant
    def is_triangle(self):
        check_a = self.a <= self.b + self.c  
        check_b = self.b <= self.a + self.c  
        check_c = self.c <= self.a + self.b  

        return "JA" if (check_a and check_b and check_c) else "NEJ"

    # 5. Skriv en method area, der udregnet trekantens areal(herons formel)
    def area(self):
        # Heron's formel - https://en.wikipedia.org/wiki/Heron%27s_formula -> s = (a+b+c)/2 -> areal = sqrt(s*(s-a)*(s-b)*(s-c))

        a, b, c = self.a, self.b, self.c
        s = (a + b + c) / 2
        return (s * (s - a) * (s - b) * (s - c)) ** 0.5

    #returnere Triangle objektets side længder
    def __str__(self):
        return f"Trekant med siderne: a: {self.a} , b: {self.b} , c: {self.c}"
    
    def find_third_point(self, x1, y1, x2, y2, a, b, c):
        # Udregn distance mellem 2 kendte punkter 
        side_a = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)       
        # Cosinus lov til at finde vinkel modsat side c
        angle_C = math.acos((a**2 + b**2 - c**2) / (2 * a * b))
        # udregn koordinater til p3 ved brug af vinkel og kendt punkt 
        x3 = round(x1 + (a * math.cos(angle_C)),2)
        y3 = round(y1 + (a * math.sin(angle_C)),2)
        return Point(x3, y3)

    # draw method der tegner trekanten via. graphics.py objekter
    def draw(self):
        if not self.is_triangle() == "JA":
            print("Ugyldig Trekant")
            return 
            
        size = self.normalizer 
        triangle_win = GraphWin("Trekant tester", 500, 500)
        triangle_win.setCoords(0, 0, size+5, size+5)
        text1 = Text(Point(2+size//10, (size+2)-size//10), f"""
        Trekantens længder: 
        a: {self.a}
        b: {self.b}
        c: {self.c}
        ______________________
        Areal: {round(self.area(),2)}
        """)
        text1.draw(triangle_win)
        # istedet for 0,0 er alle punkter skubbet med +2, for at have en lille margin i vinduet til kanterne
        p1 = Point(2 ,2)
        p2 = Point(self.a+2,2)
        p3 = self.find_third_point(2,2,self.a+2,2,self.a,self.b,self.c)
        triangle = Polygon(p1, p2, p3) 
        triangle.draw(triangle_win)
        triangle_win.getMouse()
        triangle_win.close()
````

#### - main.py

````python
# main.py

# input felter

from graphics import *
#from triangle.py Import Triangle


def main_window():
    win = GraphWin("Triangle", 300, 200)
    win.setCoords(0, 0, 20, 20)  # Sætter det som et alm. koordinatsystem , hvor (0,0) er i venstre nederste hjørne

    # Overskrift
    heading = Text(Point(10, 18), "Indtast længderne på trekanten")
    heading.draw(win)
    # info felt
    info = Text(Point(10, 16), "")
    info.draw(win)

    # Input felter 1 til 3 med labels

    # Input felt 1
    text1 = Text(Point(5, 12), "Side a")
    input1 = Entry(Point(10, 12), 10)
    input1.setFill("white")
    text1.draw(win)
    input1.draw(win)

    # Input felt 2
    text2 = Text(Point(5, 9), "Side b")
    input2 = Entry(Point(10, 9), 10)
    input2.setFill("white")
    text2.draw(win)
    input2.draw(win)

    # Input felt 3
    text3 = Text(Point(5, 6), "Side c")
    input3 = Entry(Point(10, 6), 10)
    input3.setFill("white")
    text3.draw(win)
    input3.draw(win)

    # Submit button
    submit_btn = Rectangle(Point(4, 2), Point(16, 4))
    submit_btn.draw(win)
    submit_label = Text(Point(10, 3), "Test Trekant")
    submit_label.draw(win)

    while True:

        click = win.getMouse()
        if 4 <= click.getX() <= 16 and 2 <= click.getY() <= 4:
            triangle = Triangle(input1.getText(), input2.getText(), input3.getText())
            if triangle.is_triangle() == "JA":
                try:
                    triangle.draw()
                except GraphicsError: # Den forsøger at kalde .draw hvis jeg manuelt lukker app´en. så derfor nød til try/except
                    win.close()
            else:
                info.setTextColor("red")
                info.setText("Ugyldig!")






if __name__ == "__main__":
    pass
    #main_window()
````
